```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
```

```python
X, y = load_digits(return_X_y=True)
    
X.shape
```

```python
plt.imshow(load_digits().images[7], cmap=plt.cm.Greys_r)
```

```python
tsne = TSNE(n_components=2, perplexity=30).fit(X)

Z = tsne.embedding_
```

```python
plt.scatter(Z[:, 0], Z[:, 1], c=y)
```

```python
for p in range(5, 50, 10):
    tsne = TSNE(n_components=2, perplexity=p).fit(X)
    Z = tsne.embedding_
    fig, ax = plt.subplots(1, 1)
    ax.scatter(Z[:, 0], Z[:, 1], c=y)
    ax.set_title("P = " + str(p))
```
