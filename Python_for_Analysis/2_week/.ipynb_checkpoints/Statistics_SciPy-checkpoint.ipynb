```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from scipy import stats

%matplotlib inline
```

```python
rv_bernoulli = stats.bernoulli(p = 0.4)

plt.hist(rv_bernoulli.rvs(100), density = True)
```

```python
rv_binom = stats.binom(100, p = 0.7)

rv_binom.rvs(10)
```

```python
plt.hist(rv_binom.rvs(100), bins = 10)
```

```python
a = 5
b = 10

rv_uniform = stats.uniform(a, b - a)
rv_uniform.rvs(10)
```

```python
x = np.arange(0, 15, 0.01)

plt.plot(x, rv_uniform.cdf(x))
```

```python
plt.plot(x, rv_uniform.pdf(x))
```

```python
m = 5
d = 3

rv_norm = stats.norm(m, d)

rv_norm.rvs(10)
```

```python
x = np.arange(0, 10, 0.01)

plt.plot(x, rv_norm.cdf(x))
```

```python
plt.plot(x, rv_norm.pdf(x))
```

```python
sigms = [1, 2, 3]

for s in sigms:
    rv = stats.norm(m, s)
    plt.plot(rv.pdf(x), label = str(s))

plt.legend()
```

```python
mean, scale, skew = rv_norm.stats(moments="mvs")
print(mean, scale, skew)
```
