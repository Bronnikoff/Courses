```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data/titanic.csv")
df.head()
```

```python
df["Fare"].hist(bins = 15)
```

```python
df["Fare"].plot.kde()
```

```python
df.plot.scatter(x = "Fare", y = "Survived")
```

```python
df.groupby("Survived")["Fare"].plot.kde()
plt.xlim(0, 250)
plt.legend()
```

```python
ax = df["Fare"].hist()
ax.set_title("Fare")
```

```python
fig, ax = plt.subplots(figsize = (15, 8))

df.plot.hist(y = "Fare", ax = ax)
```

```python
fig, ax = plt.subplots(figsize = (15, 8))

df["Survived"].plot.kde(label = "All", ax = ax)

for l, cls in df.groupby("Pclass"):
    print(l)
    cls.Survived.plot.kde(label = l, ax = ax)
    
plt.legend()
```
